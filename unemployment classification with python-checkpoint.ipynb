```python
data1.columns=['States','Date','Frequency','Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)','Region','longitude','latitude']
```

```python
data1['Month']=pd.to_datetime(data1['Date']).dt.month
```

```python
#data visualization
```

```python
#barplot
plt.title('unemployment estimation month wise')
sns.barplot(data=data1,x="Month",y="Estimated Unemployment Rate (%)")
plt.show()
```

```python
#barplot
plt.title('Employment estimation month wise')
sns.barplot(data=data1,x="Month",y="Estimated Employed")
plt.show()
```

```python
#boxplot
plt.title('State wise unemployment estimation')
sns.boxplot(orient='h',data=data1,x="Estimated Unemployment Rate (%)",y="States")
plt.show()
```

```python
#boxplot
plt.title('Region wise employment estimation')
sns.boxplot(orient='v',data=data1,y="Estimated Employed",x="Region")
plt.show()
```

```python
#histogram
plt.title('employed estimation month wise')
sns.histplot(x="Estimated Employed",hue="Month",data=data1)
plt.show()
```

```python
plt.title('Region wise Unemployment estimation')
sns.histplot(x="Estimated Unemployment Rate (%)",hue="Region",data=data1)
plt.show()
```
